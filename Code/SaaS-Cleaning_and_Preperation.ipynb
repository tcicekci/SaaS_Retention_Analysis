Data Loading & Cleaning

```python
import numpy as np
import pandas as pd
import datetime 
```

```python
users = pd.read_csv(r"/Users/tugrulcicekci/Desktop/Portföy Projeleri/5.SaaS/Project-5 Datasets NEW/users_new.csv")
events = pd.read_csv(r"/Users/tugrulcicekci/Desktop/Portföy Projeleri/5.SaaS/Project-5 Datasets NEW/events_new.csv")
```

```python
users.columns.strip()
users.dropna(subset=["user_id", "registration_date", "initial_plan_type",	
                    "trial_start_date",	"trial_end_date",	"first_paid_date",
                    "current_plan_type",	"churn_date",	"industry",	
                    "company_size",	"acquisition_channel"], inplace=True)
users.head()
```

```python
users.info()
```

```python
events.columns.strip()
events.dropna(subset=["event_id", "user_id", "timestamp",
                       "event_name", "session_id"], inplace=True)
events.head()
```

```python
events["timestamp"] = pd.to_datetime(events["timestamp"]).dt.date
events.head()
```

```python
events.info()
```

Data Preperation

```python
#Calculating days active by using groupby
days_active = events.groupby("user_id")["timestamp"].nunique()
users["days_active"] = users["user_id"].map(days_active)
users.head()
```

```python
events["timestamp"] = pd.to_datetime(events["timestamp"])
events["week_of_year"] = events["timestamp"].dt.isocalendar().week
events["month_of_year"] = events["timestamp"].dt.month
events.head()
```

```python
first_active = events.groupby("user_id")["timestamp"].min().reset_index()
first_active["first_active_week"] = first_active["timestamp"].dt.isocalendar().week
first_active["first_active_month"] = first_active["timestamp"].dt.month
users = users.merge(first_active,
                    on="user_id",
                    how="left")
users.head()
```

```python
users.drop(columns="timestamp", inplace=True)
users.head()
```

```python
users.to_csv("cleaned_users.csv")
events.to_csv("cleaned_events.csv")
```
